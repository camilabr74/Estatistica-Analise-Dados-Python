```
import numpy as np
import pandas as pd
from collections import Counter
```

```
enem_sp = pd.read_csv('/content/drive/MyDrive/programa_formacao/Estatistica_Analise_Dados_Python/enem_2019_tratado.csv',
                      sep=',', encoding='iso-8859-1')
```

```
enem_sp.head()
```

```
enem_sorocaba = enem_sp.loc[enem_sp.NO_MUNICIPIO_RESIDENCIA == 'Sorocaba']
```

```
enem_sorocaba.head()
```

```
enem_sorocaba.shape
```

Localizando idades

```
idades = enem_sorocaba['IDADE']
```

```
idades
```

Counter para contar frequência absoluta. - quantas vezes um registro se repetiu.

```
freq_absoluta = Counter(idades)
```

```
freq_absoluta
```

```
freq_absoluta = pd.DataFrame.from_dict(freq_absoluta, orient='index')
```

```
freq_absoluta
```

```
freq_absoluta = freq_absoluta.sort_index(ascending=True)
freq_absoluta
```

```
freq_relativa = freq_absoluta / freq_absoluta.sum()
freq_relativa
```

```
freq_relativa_perc = freq_relativa * 100
freq_relativa_perc
```
